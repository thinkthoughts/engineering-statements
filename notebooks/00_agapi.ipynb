## 2. Define the AGAPI Engineering Statement

In the repository, this should live at:

```text
statements/agapi-agents.yaml
```

For Colab portability, the YAML is embedded here too.


````python
agapi_yaml = 'id: agapi-agents\ntitle: Engineering Statements With AGAPI-Agents\n\nsource:\n  paper: https://arxiv.org/abs/2512.11935\n  github: https://github.com/atomgptlab/agapi\n  feature: https://climatereality.app/statements/\n  repo: https://github.com/thinkthoughts/engineering-statements\n\nobjective: >\n  Connect AGAPI-style agentic scientific infrastructure with portable\n  Engineering Statements that specify claims, constraints, evidence,\n  workflows, outputs, and next steps.\n\ncontexts:\n  - agentic science\n  - materials discovery\n  - open scientific infrastructure\n  - scientific reasoning workflows\n  - reproducible research artifacts\n\nconstraints:\n  - preserve scientific intent\n  - support collaboration rather than centralization\n  - keep outputs inspectable and reproducible\n  - complement existing scientific tools and repositories\n\nevidence:\n  - AGAPI paper\n  - AGAPI GitHub repository\n  - Climate Reality statements feature\n  - Engineering Statements examples derived from recent papers\n\noutputs:\n  - feature page\n  - YAML statements\n  - companion notebooks\n  - lab reports\n  - reusable templates\n\nnext_steps:\n  - read the AGAPI paper\n  - inspect the AGAPI repository\n  - compare AGAPI workflows with Engineering Statements\n  - explore how Engineering Statements might complement agentic science systems\n  - provide feedback on the specification structure\n'

print(agapi_yaml)
````

## 3. Load the YAML

YAML lets humans and tools read the same specification.


````python
import yaml
from pathlib import Path
import json

statement = yaml.safe_load(agapi_yaml)
statement
````

## 4. Render a readable Engineering Statement

This turns the YAML into a compact text summary.


````python
def render_statement_md(statement: dict) -> str:
    lines = []
    lines.append(f"# {statement['title']}")
    lines.append("")
    lines.append(f"**ID:** `{statement['id']}`")
    lines.append("")
    lines.append("## Objective")
    lines.append(statement["objective"].strip())
    lines.append("")
    lines.append("## Sources")
    for key, value in statement.get("source", {}).items():
        lines.append(f"- **{key}:** {value}")
    lines.append("")
    for section in ["contexts", "constraints", "evidence", "outputs", "next_steps"]:
        title = section.replace("_", " ").title()
        lines.append(f"## {title}")
        for item in statement.get(section, []):
            lines.append(f"- {item}")
        lines.append("")
    return "\n".join(lines)

statement_md = render_statement_md(statement)
print(statement_md)
````

## 5. Save reusable artifacts

This writes:

```text
agapi-agents.yaml
agapi-agents.json
agapi-agents.md
```

These can be used by future scripts, lab reports, or website generators.


````python
out_dir = Path("outputs")
out_dir.mkdir(exist_ok=True)

(out_dir / "agapi-agents.yaml").write_text(agapi_yaml.strip() + "\n", encoding="utf-8")
(out_dir / "agapi-agents.json").write_text(json.dumps(statement, indent=2), encoding="utf-8")
(out_dir / "agapi-agents.md").write_text(statement_md, encoding="utf-8")

print("Wrote:")
for path in sorted(out_dir.glob("agapi-agents.*")):
    print("-", path)
````
