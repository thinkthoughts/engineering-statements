## 3. Toy Example: Pairwise Sums

This first toy example does not model the paper's theorem. It simply makes the phrase "image set" concrete.

For a finite set \(A\), the image of the polynomial \(f(x,y)=x+y\) on \(A \times A\) is the set of pairwise sums.


```python
def image_set(A, f):
    return sorted({f(x, y) for x in A for y in A})

A = [2, 3, 5, 7, 11, 13]
sums = image_set(A, lambda x, y: x + y)

print("A =", A)
print("|A| =", len(A))
print("Image of x+y on A x A:")
print(sums)
print("|image| =", len(sums))
```

## 4. Toy Example: Modular Residues

The paper's proof uses local restrictions modulo primes. Here we inspect a simple background fact: primes greater than 5 live in the mod30 residue lanes that are coprime to 2, 3, and 5.

This is useful context, not the central claim of the paper.


```python
def is_prime(n):
    if n < 2:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    d = 3
    while d * d <= n:
        if n % d == 0:
            return False
        d += 2
    return True

primes = [n for n in range(2, 200) if is_prime(n)]
residues_mod30 = sorted({p % 30 for p in primes if p > 5})

print("Primes under 200:")
print(primes)
print("\nResidues mod 30 for primes > 5:")
print(residues_mod30)
```

```python
from collections import Counter

counts = Counter(p % 30 for p in primes if p > 5)

print("Counts by mod30 lane for primes under 200:")
for r in [1, 7, 11, 13, 17, 19, 23, 29]:
    print(f"{r:2d}: {counts[r]}")
```

## 5. Toy Example: Squares Modulo an Odd Prime

A key idea in the paper is that certain polynomial values are forced into square classes modulo primes.

For a small odd prime \(p\), the squares modulo \(p\), including zero, form a subset of the residue classes.


```python
def squares_mod(p):
    return sorted({(x * x) % p for x in range(p)})

for p in [3, 5, 7, 11, 13]:
    sq = squares_mod(p)
    print(f"p={p:2d} squares including 0: {sq}  count={len(sq)} of {p}")
```

## 6. Toy Polynomial With a Square-Class Restriction

The paper uses a more sophisticated construction. This toy example only illustrates the local mechanism:

If a polynomial reduces to a square modulo \(p\), its image modulo \(p\) lies inside the square classes.


```python
def toy_poly(x, y, p):
    # A simple toy form whose value mod p is a square of x+y.
    return ((x + y) ** 2) % p

p = 11
all_residues = sorted({toy_poly(x, y, p) for x in range(p) for y in range(p)})

print("Toy polynomial image residues mod", p)
print(all_residues)
print("Squares mod", p)
print(squares_mod(p))
print("Matches square classes?", all_residues == squares_mod(p))
```
