## 5. Engineering Statement Renderer

This renders the YAML into Markdown for reuse in a README, lab report, or website section.


````python
def render_statement_md(statement: dict) -> str:
    lines = []
    lines.append(f"# {statement['title']}")
    lines.append("")
    lines.append(f"**ID:** `{statement['id']}`")
    lines.append("")
    lines.append("## Objective")
    lines.append(statement["objective"].strip())
    lines.append("")
    lines.append("## Sources")
    for key, value in statement.get("source", {}).items():
        lines.append(f"- **{key}:** {value}")
    lines.append("")
    for section in ["contexts", "constraints", "evidence", "outputs", "next_steps"]:
        title = section.replace("_", " ").title()
        lines.append(f"## {title}")
        for item in statement.get(section, []):
            lines.append(f"- {item}")
        lines.append("")
    return "\n".join(lines)

statement_md = render_statement_md(statement)
print(statement_md)
````
