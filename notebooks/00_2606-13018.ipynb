## 2. Vocabulary

- **Semiprime:** an integer \(N=pq\), usually the product of two primes.
- **Finite field \(F_\ell\):** integers modulo an odd prime \(\ell\).
- **Legendre symbol \(\chi(x)\):** tells whether \(x\) is a quadratic residue, nonresidue, or zero modulo \(\ell\).
- **Hyperbolic sieve set:** \(H_a(N;\ell)=\{ax+Nx^{-1}:x\in F_\ell^*\}\).
- **Two-target union:** \(U_{a,b}=H_a(N;\ell)\cup H_b(N;\ell)\).
- **Character sum:** a finite-field sum involving \(\chi\), here controlling the size of \(U_{a,b}\).
- **Frobenius trace:** \(t_E=\ell+1-\#E(F_\ell)\), measuring how many points an elliptic curve has over a finite field.
- **CRT product:** combining local residue information modulo several primes through the Chinese remainder theorem.


## 3. Finite-Field Helpers

These helpers implement modular inverses, Legendre symbols, and hyperbolic sieve sets for small examples.


```python
def inv_mod(x, ell):
    return pow(x % ell, -1, ell)

def legendre(a, ell):
    a %= ell
    if a == 0:
        return 0
    value = pow(a, (ell - 1) // 2, ell)
    return -1 if value == ell - 1 else value

def H(a, N, ell):
    return sorted({(a * x + N * inv_mod(x, ell)) % ell for x in range(1, ell)})

def union_size(a, b, N, ell):
    Ha = set(H(a, N, ell))
    Hb = set(H(b, N, ell))
    return len(Ha | Hb), sorted(Ha | Hb)

def K(A, B, ell):
    return sum(legendre((z*z - A) * (z*z - B), ell) for z in range(ell))
```

## 4. Toy Example: One-Target Sieve Size

The paper uses the exact formula:

\[
|H_a(N;\ell)| = \frac{\ell + \chi(aN)}{2}.
\]

This cell checks the formula for a small prime.


```python
ell = 11
N = 6
a = 3

Ha = H(a, N, ell)
formula = (ell + legendre(a * N, ell)) // 2

print("ell =", ell, "N =", N, "a =", a)
print("H_a =", Ha)
print("|H_a| =", len(Ha))
print("Formula =", formula)
```

## 5. Toy Example: Two-Target Union

For two coefficients \(a\neq b\), the paper shows the union size is controlled by the character sum

\[
K(A,B)=\sum_{z\in F_\ell}\chi((z^2-A)(z^2-B)),
\]

where \(A=4aN\) and \(B=4bN\).


```python
ell = 13
N = 1
a = 7
b = (a * a) % ell

A = (4 * a * N) % ell
B = (4 * b * N) % ell

u_size, U = union_size(a, b, N, ell)
k_value = K(A, B, ell)

print("ell =", ell, "N =", N, "a =", a, "b =", b)
print("A =", A, "B =", B)
print("H_a =", H(a, N, ell))
print("H_b =", H(b, N, ell))
print("Union =", U)
print("|Union| =", u_size)
print("3ell/4 =", 3 * ell / 4)
print("K(A,B) =", k_value)
```

## 6. Counterexample Check: The Small \(\ell=5\) Case

The paper notes a small counterexample to the pointwise benchmark \(|U_{\gamma,\gamma^2}|\le 3\ell/4+1\) at \(\ell=5, N=1, \gamma=2\).

This cell reproduces the local set computation.


```python
ell = 5
N = 1
gamma = 2
a = gamma
b = (gamma * gamma) % ell

u_size, U = union_size(a, b, N, ell)

print("ell =", ell, "N =", N, "gamma =", gamma, "gamma^2 =", b)
print("H_gamma =", H(a, N, ell))
print("H_gamma^2 =", H(b, N, ell))
print("Union =", U)
print("|Union| =", u_size)
print("Benchmark 3ell/4 + 1 =", 3 * ell / 4 + 1)
print("Benchmark fails?", u_size > 3 * ell / 4 + 1)
```

## 7. Character Sum to Curve Context

For \(A\neq B\), the degree-four expression

\[
(z^2-A)(z^2-B)
\]

defines the affine curve

\[
Y^2=(X^2-A)(X^2-B).
\]

The paper connects \(K(A,B)\) to the Frobenius trace of the smooth projective genus-one curve associated with this equation. The Hasse–Weil bound then gives an \(O(\sqrt{\ell})\) error scale.

This notebook does not construct the projective curve. It keeps the transition visible:

\[
\text{two-target union size}
\rightarrow
\text{character sum}
\rightarrow
\text{elliptic trace obstruction}.
\]


```python
def scan_two_target_examples(ell, N):
    rows = []
    for a in range(1, ell):
        for b in range(1, ell):
            if a == b:
                continue
            A = (4 * a * N) % ell
            B = (4 * b * N) % ell
            u_size, _ = union_size(a, b, N, ell)
            rows.append({
                "a": a,
                "b": b,
                "A": A,
                "B": B,
                "union_size": u_size,
                "deviation_from_3ell4": u_size - 3 * ell / 4,
                "K": K(A, B, ell),
            })
    return rows

rows = scan_two_target_examples(11, 1)
rows[:10]
```
