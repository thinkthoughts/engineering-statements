## 2. Vocabulary

- **q-series:** a power series in a variable \(q\), often encoding combinatorial or modular information.
- **Nahm sum:** a q-hypergeometric series associated with quadratic data.
- **Modularity:** transformation behavior under modular substitutions.
- **Modular triple:** data that produces a modular Nahm sum.
- **Rogers–Ramanujan identity:** an identity expressing a q-series as an infinite product.
- **Bailey pair:** a tool for deriving q-series identities.
- **Lift-dual operation:** a construction that can generate higher-rank examples from known modular data.


## 3. Toy Example: A Simple q-Series

This example does not model a Nahm sum. It only introduces the idea that q-series can be inspected numerically through partial sums.

For \(|q| < 1\):

\[
1 + q + q^2 + q^3 + \cdots = \frac{1}{1-q}.
\]


```python
def partial_geometric(q, N):
    return sum(q**n for n in range(N))

for q in [0.1, 0.25, 0.5, 0.8]:
    approx = partial_geometric(q, 30)
    exact = 1 / (1 - q)
    print(f"q={q:0.2f} partial={approx:0.8f} exact={exact:0.8f} error={abs(approx-exact):0.2e}")
```

## 4. Toy Example: Partition Numbers

Rogers–Ramanujan identities are closely related to partitions and product expansions.

This cell computes small partition numbers \(p(n)\), where \(p(n)\) is the number of ways to write \(n\) as a sum of positive integers, ignoring order.


```python
def partition_numbers(n_max):
    p = [0] * (n_max + 1)
    p[0] = 1
    for k in range(1, n_max + 1):
        for n in range(k, n_max + 1):
            p[n] += p[n - k]
    return p

p = partition_numbers(20)

for n, val in enumerate(p):
    print(f"p({n:2d}) = {val}")
```

## 5. Toy Example: Infinite Products

Many q-series identities equate a sum-side expression with a product-side expression.

This example computes a finite truncation of a simple infinite product:

\[
\prod_{k=1}^{N} \frac{1}{1-q^k}.
\]

This product is the generating function for partition numbers as \(N\) grows.


```python
def partition_product(q, N):
    product = 1.0
    for k in range(1, N + 1):
        product *= 1 / (1 - q**k)
    return product

for q in [0.1, 0.25, 0.5]:
    print(f"q={q}")
    for N in [5, 10, 20, 40]:
        print(f"  N={N:2d}: {partition_product(q, N):0.8f}")
```
